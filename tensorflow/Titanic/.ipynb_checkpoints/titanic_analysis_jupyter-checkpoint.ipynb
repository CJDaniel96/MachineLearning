```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('titanic//train.csv')
test = pd.read_csv('titanic//test.csv')
submit = pd.read_csv('titanic//gender_submission.csv')
```

```python
data = train.append(test)
```

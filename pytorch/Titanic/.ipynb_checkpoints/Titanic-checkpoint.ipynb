```python
class TitanicDataset(Dataset):
    def __init__(self, dataframe, mode='train'):
        if mode == 'train':
            train_df = dataframe[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
            train_df['Age'].fillna(train_df['Age'].mean(), inplace=True)
            train_df['Embarked'].fillna(train_df['Embarked'].value_counts().idxmax(), inplace=True)
            train_df['Sex'] = train_df['Sex'].map({
                'male':0,
                'female':1
            })
            train_df['Embarked'] = train_df['Embarked'].map({
                'S':0,
                'C':1,
                'Q':2
            })
            target = dataframe['Survived']
            
            self.x = torch.from_numpy(train_df.values)
            self.y = torch.from_numpy(target.values)
            
        elif mode == 'test':
            test_df = dataframe[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
            test_df['Age'].fillna(test_df['Age'].mean(), inplace=True)
            test_df['Fare'].fillna(test_df['Fare'].mean(), inplace=True)
            
            self.x = torch.from_numpy(test_df.values)
            
    def __getitem__(self, index):
        return self.x[index], self.y[index]
    
    def __len__(self):
        return self.x.shape[0]
```
